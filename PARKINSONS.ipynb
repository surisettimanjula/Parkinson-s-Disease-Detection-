#Extending dataset


```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE

# 1. Load your original dataset
df = pd.read_csv("/content/parkinsons.csv")

# 2. Separate features and target
X = df.drop(columns=['name', 'status'])
y = df['status']

# 3. Apply SMOTE to balance classes
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# 4. Convert to DataFrame
X_smote_df = pd.DataFrame(X_smote, columns=X.columns)
y_smote_df = pd.Series(y_smote, name='status')

# 5. Add noise to create synthetic diversity (Gaussian noise)
def add_noise(data, noise_level=0.01):
    noisy_data = data.copy()
    noise = np.random.normal(0, noise_level, size=data.shape)
    return noisy_data + noise

X_extra = add_noise(X_smote_df, noise_level=0.02)
y_extra = y_smote_df.copy()

# 6. Combine original + SMOTE + noisy extra
X_all = pd.concat([X, X_smote_df, X_extra], ignore_index=True)
y_all = pd.concat([y, y_smote_df, y_extra], ignore_index=True)

# 7. Re-attach the 'name' column (optional fake names)
name_column = ['synthetic_' + str(i) for i in range(len(X_all))]
df_extended = pd.DataFrame(X_all)
df_extended['name'] = name_column
df_extended['status'] = y_all

# 8. Reorder columns to match original format
cols = ['name'] + [col for col in df_extended.columns if col not in ['name', 'status']] + ['status']
df_extended = df_extended[cols]

# 9. Save extended dataset
df_extended.to_csv("/content/parkinsons_extended.csv", index=False)
print("✅ Dataset extended and saved as 'parkinsons_extended.csv'")
```

```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Bidirectional, GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping

# Step 2: Load Extended Dataset
df = pd.read_csv("/content/parkinsons_extended.csv")
print("Dataset shape:", df.shape)

X = df.drop(columns=['name', 'status'])
y = df['status']

# Step 3: Scale and Reshape
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_reshaped = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_reshaped, y, test_size=0.2, random_state=42, stratify=y
)
print("Train samples:", X_train.shape[0], "| Test samples:", X_test.shape[0])

# Step 5: Define CNN + BiGRU Model
def build_model(input_shape):
    model = Sequential([
        Conv1D(32, kernel_size=3, activation='relu', input_shape=input_shape, kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),
        Bidirectional(GRU(32, return_sequences=False, kernel_regularizer=l2(0.001))),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu', kernel_regularizer=l2(0.001)),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Step 6: Train Model with EarlyStopping
model = build_model((X_train.shape[1], 1))
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)

# Step 7: Evaluate on Test Set
y_pred = model.predict(X_test)
y_pred_class = (y_pred > 0.5).astype(int)

test_accuracy = accuracy_score(y_test, y_pred_class)
print("Test Accuracy:", round(test_accuracy, 4))

print("\nClassification Report:\n", classification_report(y_test, y_pred_class))

cm = confusion_matrix(y_test, y_pred_class)
print("Confusion Matrix:\n", cm)

# Step 8: Cross-Validation (5-Fold)
print("\nRunning 5-Fold Cross-Validation...")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for train_idx, val_idx in kf.split(X_reshaped):
    X_cv_train, X_cv_val = X_reshaped[train_idx], X_reshaped[val_idx]
    y_cv_train, y_cv_val = y.iloc[train_idx], y.iloc[val_idx]

    temp_model = build_model((X_cv_train.shape[1], 1))
    temp_model.fit(X_cv_train, y_cv_train, epochs=20, batch_size=16, verbose=0)

    y_cv_pred = temp_model.predict(X_cv_val)
    y_cv_class = (y_cv_pred > 0.5).astype(int)
    acc = accuracy_score(y_cv_val, y_cv_class)
    cv_scores.append(acc)

print("Cross-validation accuracies:", cv_scores)
print("Mean CV Accuracy:", round(np.mean(cv_scores), 4))
print("Standard Deviation:", round(np.std(cv_scores), 4))

# Step 9: Plot Accuracy and Loss Graphs
plt.figure(figsize=(10, 4))
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 10: Plot Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Healthy (0)', 'Parkinson’s (1)'],
            yticklabels=['Healthy (0)', 'Parkinson’s (1)'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```
